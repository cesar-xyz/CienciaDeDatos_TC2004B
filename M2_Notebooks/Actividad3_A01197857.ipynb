## Modelo Entidad-Relación.

#### ¿Por qué es importante los modelos de datos en la fase de integración?
    
   Al tener datos masivos es normal que las empresas y organizaciones tengan un desastre en su integración de datos por la masividad de estos, una de las unicas formas de resolver esta problematica es el uso de los **modelos de datos** ya que ayudan a describir la manipulación de los datos que se utilizaran en la empresa y esto ayuda a la integración de los datos.
   
#### ¿Qué es un Modelo Entidad-Relación? ¿El Modelo Entidad-Relación es un modelo Conceptual, Lógico o Físico?

   Un modelo de entidad relación es modelo conceptual, el cual ayuda a las empresas a visualizar el flujo de datos por medio de información de ellas, identifica entidades y su relación de ellas. Propicia a que la integración de los datos sea más facil y tener un diseño visual de lo que se quiere realizar.
   
#### ¿Cuáles son los elementos principales en un modelo Entidad-Relación?

   - **Entidad**. Se pueden identificar datos de interés y que se quieren registrar.
   - **Atributo**. Describe las características de un individuo, incluidas las simples, complejas, multivaluadas y derivadas.
   - **Identificador**. También conocida como "llave", identifica una instancia de entidad.
   - **Relaciones, participación y notaciones**.

### Modeló Entidad-Relación aeropuerto.

En un aeropuerto se requiere una base de datos para guardar la información de todos sus aviones, y los empleados del aeropuerto. De los requerimientos para esta base de datos se colectó la siguiente información:


Cada avión tiene un número único de registro (#Reg), un nombre y es de un modelo de avión particular, por ejemplo modelo Boeing 777. De cada modelo de avión se registra el número de modelo, su capacidad (número de pasajeros que puede transportar) y su velocidad. Cada avión es almacenado en un hangar particular, del cual se registra su número, capacidad y localización. En un hangar se pueden almacenar varios aviones.


Se tiene además un registro de los pilotos. De cada piloto se guarda su número de licencia, nombre, edad, modelos de avión que está autorizado  manejar y la fecha en la que empezó a manejar cada uno de esos modelos de avión.


Un avión puede realizar varios vuelos. De cada vuelo se registra el número de vuelo, la ciudad origen y la ciudad destino. De cada ciudad se registra su clave y nombre. 


Un avión pertenece a una compañía, de las cuales se registra su RFC, nombre, dirección y número de teléfono y la fecha de creación de la misma. Una compañía puede ser dueña de varios aviones.


El aeropuerto tiene un conjunto de empleados de mantenimiento, de los cuales se registra su número único, nombre completo (nombre, apellido), sexo, número de IMSS, salario y edad.


Sobre un avión pueden realizarse varios servicios de mantenimiento numerados consecutivamente por avión. Del servicio se registra la fecha en que fue realizado, el trabajo realizado y la duración en horas. Por ejemplo: al avión D1 se le han hecho dos servicios, el primero, fue la semana pasada y duro 5 horas, y el segundo fue hoy y duró 2 horas; al avión D2 se le hizo ayer solo un servicio que duró 3 horas. Un servicio puede ser llevado a cabo por varios empleados.
