# Preprocesamiento de datos

```python
# Quitar Warnings.
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
```

```python
# Importar librerias
import pandas as pd
import numpy as np
```

```python
# Importar el conjunto de datos 'Cars1'
cars1 = pd.read_csv('Cars1.csv')
cars1
```

```python
# Importar el conjunto de datos 'Cars2'
cars2 = pd.read_csv('Cars2.csv')
cars2
```

**Comparación de los archivos**

Los dos archivos tienen la misma estructura, aunque el conjunto de datos 'Cars1' tiene dos columnas adicionales: 'data1' y 'data2'. Estos conjuntos de datos no cuentan con una columna igual por la cual puedan unirse. La solución es unir un conjunto de datos a otro como filas nuevas del conjunto de datos final.

```python
# Unir bases de datos
data = cars1.append(cars2)
data
```

```python
# Remover columnas que contienen sólo valores NaN
del(data['data1'], data['data2'])

# Comprobar 
data
```

```python
# Buscar los valores nulos '?' y reemplazarlos por 'NaN'
data = data.replace('?',np.nan)
```

```python
# Pasar las variables de 'object' a 'int64' para sacar las medias
data['horsepower'] = pd.to_numeric(data['horsepower']) 
data['acceleration'] = pd.to_numeric(data['acceleration']) 
```

```python
# Reemplazar los valores NaN por la media correspondiente
data.fillna(data.mean())
```

```python
# Mostrar resumen
display(data.describe())
display(data.info())
```
