# 0. Comenzaremos con la importación de las bibliotecas de Python requeridas.

```python
import numpy as np
import matplotlib.pyplot as plt
import keras
import warnings
from mpl_toolkits import mplot3d
from matplotlib import cm
from keras.models import Sequential
from keras.layers import Dense

# from keras.optimizers import Adam

from tensorflow import keras
from tensorflow.keras.optimizers import Adam

from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
warnings.filterwarnings("ignore")
```

# 1. Crea un conjunto de datos artificial

## 1.1 Arreglar semilla aleatoria para reproducibilidad

```python
seed = 7
np.random.seed(seed)
```

## 1.2 Necesitamos generar dos conjuntos de datos

```python
# The first way is by using the already existing functions:
X,y = make_circles(n_samples = 3000, noise = 0.08, factor=0.3)
print(X.shape)
print(y.shape)
print(type(X))
print(type(y))

# Spliting data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# The other way is by hand:
X0 = [X[i,:] for i in range(len(y)) if y[i] == 0] # returns the row where the corresponding label is 0
X1 = [X[i,:] for i in range(len(y)) if y[i] == 1] # returns the row where the corresponding label is 1
            
X0_np = np.array(X0) # convert it into an array
X1_np = np.array(X1) 

X0_train = X0_np[:1000,:].T # picked the first 1000 samples in the second axis
X0_test = X0_np[1000:,:].T # picked the remaining 500 samples in the second axis

X1_train = X1_np[:1000,:].T  # picked the first 1000 samples in the second axis
X1_test = X1_np[1000:,:].T # picked the remaining 500 samples in the second axis

X_train = np.hstack([X0_train,X1_train]).T # all training examples
y_train=np.zeros((1,2000))
y_train[0, 1000:] = 1

X_test = np.hstack([X0_test,X1_test]).T # all test examples
y_test=np.zeros((1,1000))
y_test[0, 500:] = 1
```

## 1.3 Visualización de los datos

```python
plt.figure(figsize=(12,8))
plt.scatter(X0_train[0,:],X0_train[1,:], color = 'b', label = 'class 0 train')
plt.scatter(X1_train[0,:],X1_train[1,:], color = 'r',  label = 'class 1 train')
plt.scatter(X0_test[0,:],X0_test[1,:], color = 'LightBlue', label = 'class 0 test')
plt.scatter(X1_test[0,:],X1_test[1,:], color = 'Orange', label = 'class 1 test')
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.legend()
plt.axis('equal')
plt.show()
```

## 1.4 Comprobación de la forma de los datos de entrada

```python
print('x_train:\t{}' .format(X_train.shape))
print('y_train:\t{}' .format(y_train.shape))
print('x_test:\t\t{}'.format(X_test.shape))
print('y_test:\t\t{}'.format(y_test.shape))
```

# 2. Crea la red

## 2.1 Hiperparámetros

```python
# Hyper-parameters
training_epochs = 400 # Total number of training epochs
learning_rate = 0.03 # The learning rate
```

## 2.1 Creando un modelo
Nuestro modelo tendrá 2 capas, con 6 neuronas en la capa de entrada y una neurona en la capa de salida. Usaremos un inicializador normal que genera tensores con una distribución normal.

El optimizador que usaremos es Adam. Es un algoritmo de optimización que se puede usar en lugar del procedimiento de descenso de gradiente estocástico clásico para actualizar iterativamente los pesos de la red en función de los datos de entrenamiento. Adam es un algoritmo popular en el campo del aprendizaje profundo porque logra buenos resultados rápidamente. Los parámetros predeterminados siguen los proporcionados en el documento original.

Para que esto funcione en keras, necesitamos compilar el modelo. Una elección importante que hacer es la función de pérdida. Usamos la pérdida de entropía cruzada binaria y no la habitual en la clasificación multiclase utilizada pérdida de entropía cruzada categórica. Esto puede parecer poco razonable, pero queremos penalizar cada nodo de salida de forma independiente. Así que elegimos una pérdida binaria y modelamos la salida de la red como distribuciones de Bernoulli independientes por etiqueta.

```python
# create model
def create_model():
    model = Sequential()
    
    # Input layer
    model.add(Dense(6, input_dim=2, kernel_initializer='normal', activation='relu'))
    
    # Output layer
    model.add(Dense(y_train.T.shape[1], activation='sigmoid'))

    # Compile model
    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate), metrics = ['accuracy'])
    return model

model = create_model()
model.summary()
```

## 2.2 Entrenar el modelo
Entrenemos el modelo para un número determinado de épocas.

```python
results = model.fit(
    X_train, y_train.T,
    epochs= training_epochs,
    batch_size = 516,
    validation_data = (X_test, y_test.T),
    verbose = 0
)
```

## 2.3 Probar el modelo
El modelo puede generar predicciones de salida para las muestras de entrada.

```python
prediction_values = model.predict(X_test)
```

## 2.4 Valor de error medio

```python
print("Test-Accuracy:","%.2f%%" % (np.mean(results.history["val_accuracy"])*100))
```

## 2.5 Evaluar el modelo para ver la precisión
Ahora podemos comprobar la precisión de nuestro modelo.

```python
print("Evaluating on training set...")
(loss, accuracy) = model.evaluate(X_train, y_train.T)
print("loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))

print("Evaluating on testing set...")
(loss, accuracy) = model.evaluate(X_test, y_test.T)
print("loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))
```

## 2.6 Resumir el historial para mayor precisión


```python
plt.plot(results.history['accuracy'])
plt.plot(results.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='down right')
```

## 2.7 Resumir el historial de pérdida

```python
plt.plot(results.history['loss'])
plt.plot(results.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')

max_loss = np.max(results.history['loss'])
min_loss = np.min(results.history['loss'])
print("Maximum Loss : {:.4f}".format(max_loss))
print("")
print("Minimum Loss : {:.4f}".format(min_loss))
print("")
print("Loss difference : {:.4f}".format((max_loss - min_loss)))
```

## 3.3 Mostrar los pesos y sesgos de nuestro modelo

```python
# Input layer
weights0 = model.layers[0].get_weights()[0]
biases0 = model.layers[0].get_weights()[1]
print("Input layer weights",weights0.shape,":\n",weights0)
print("Input layer biases",biases0.shape,":\n",biases0)

# Output layer
weights1 = model.layers[1].get_weights()[0]
biases1 = model.layers[1].get_weights()[1]
print("\nOutput layer weights",weights1.shape,":\n",weights1)
print("Output layer biases",biases1.shape,":\n",biases1)
```

## 3.4 Funciones de activación
ReLu: la unidad lineal rectificada es la opción predeterminada de funciones de activación en la capa oculta. En la capa de salida, usamos Sigmoid como función de activación, porque su salida está en el rango entre 0 y 1.

Al usar ReLu en la capa oculta, la red neuronal aprenderá mucho más rápido que sigmoid o tanah, porque la pendiente de sigmoid y tanh será 0 si z es un número positivo o negativo grande y ralentiza el descenso del gradiente. La derivada de ReLu es 1 si z>0.

Primero debemos definirlos:

```python
def sigmoid(z):
    return 1./(1+np.exp(-z))

def ReLU(z):
    return np.maximum(0,z)
  
z = np.linspace(-10,10,100)
plt.plot(z, sigmoid(z),'r', label = 'sigmoid')
plt.plot(z, ReLU(z),'b',label ='ReLU')
plt.legend(fontsize=12)
```

## 3.5 Calcular la salida de cada neurona en nuestra red
Calcule la salida de cada neurona en nuestra red, paso a paso.

Para obtener más información sobre las redes neuronales, puede visitar http://datahacker.rs/neural_networks_representation/.

```python
x1 = np.linspace(-2,2,50)
x2 = np.linspace(-2,2,50)
x3 = np.linspace(-1,1,50*50)

result = np.zeros((50*50,3)) 

out11 = np.zeros((50*50, 1))
out12 = np.zeros((50*50, 1))
out13 = np.zeros((50*50, 1))
out14 = np.zeros((50*50, 1))
out15 = np.zeros((50*50, 1))
out16 = np.zeros((50*50, 1))

out21 = np.zeros((50*50, 1))


for i in range(len(x1)):
    for j in range(len(x2 )):
        # Input layer
        n11 = x1[i] * weights0[0][0] + x2[j] * weights0[1][0] + biases0[0]
        a11 = ReLU(n11)

        n12 = x1[i] * weights0[0][1] + x2[j] * weights0[1][1] + biases0[1]
        a12 = ReLU(n12)

        n13 = x1[i] * weights0[0][2] + x2[j] * weights0[1][2] + biases0[2]
        a13 = ReLU(n13)

        n14 = x1[i] * weights0[0][3] + x2[j] * weights0[1][3] + biases0[3]
        a14 = ReLU(n14)

        n15 = x1[i] * weights0[0][4] + x2[j] * weights0[1][4] + biases0[4]
        a15 = ReLU(n15)

        n16 = x1[i] * weights0[0][5] + x2[j] * weights0[1][5] + biases0[5]
        a16 = ReLU(n16)

        # Output layer
        n21 = a11 *weights1[0][0] + a12 *weights1[1][0] + a13 * weights1[2][0] + \
        a14 *weights1[3][0] + a15 *weights1[4][0] +a16  *weights1[5][0] + biases1[0]
        a21 = sigmoid(n21)


        result[i*50+j,0 ] = x1[i]
        result[i*50+j, 1] = x2[j]

        out11[i*50+j] = a11
        out12[i*50+j] = a12
        out13[i*50+j] = a13
        out14[i*50+j] = a14
        out15[i*50+j] = a15
        out16[i*50+j] = a16

        out21[i*50+j] = a21
```

## 3.6 Calcular la salida de cada neurona en nuestra red
Usando el principio de vectorización, podemos simplificar las ecuaciones anteriores:

```python
x1 = np.linspace(-2,2,50)
x2 = np.linspace(-2,2,50)
x3 = np.linspace(-1,1,50*50)

result = np.zeros((50*50,3)) 

out11 = np.zeros((50*50, 1))
out12 = np.zeros((50*50, 1))
out13 = np.zeros((50*50, 1))
out14 = np.zeros((50*50, 1))
out15 = np.zeros((50*50, 1))
out16 = np.zeros((50*50, 1))

out21 = np.zeros((50*50, 1))

for i in range(len(x1)):
    for j in range(len(x2 )):
        # Input layer
        n1 = np.dot(weights0[0],x1[i]) + np.dot(weights0[1],x1[j]) + biases0
        a1 = ReLU(n1)
        # Output layer
        n2 = np.dot(weights1[:,0],a1) + biases1
        a2 = sigmoid(n2)

        result[i*50+j,0 ] = x1[i]
        result[i*50+j, 1] = x2[j]

        out11[i*50+j] = a1[0]
        out12[i*50+j] = a1[1]
        out13[i*50+j] = a1[2]
        out14[i*50+j] = a1[3]
        out15[i*50+j] = a1[4]
        out16[i*50+j] = a1[5]

        out21[i*50+j] = a2
```

## 3.7 Graficar la salida de cada neurona en la capa de entrada
Ahora podemos trazar la salida de cada neurona en la capa de entrada

```python
plt.figure(figsize =(8,8))

# First neuron, input layer
plt.subplot(231)
plt.title('First neuron hidden layer 1')
plt.scatter(result[:, 0], result[:, 1], c= out11.reshape(-1), cmap=cm.coolwarm)

# Second neuron, input layer
plt.subplot(232)
plt.title('Second neuron hidden layer 1')
plt.scatter(result[:, 0], result[:, 1], c= out12.reshape(-1), cmap=cm.coolwarm)

# Third neuron, input layer
plt.subplot(233)
plt.title('Third neuron hidden layer 1')
plt.scatter(result[:, 0], result[:, 1], c= out13.reshape(-1), cmap=cm.coolwarm)

# Fourth neuron, input layer
plt.subplot(234)
plt.title('Fourth neuron hidden layer 1')
plt.scatter(result[:, 0], result[:, 1], c= out14.reshape(-1), cmap=cm.coolwarm)

# Fifth neuron, input layer
plt.subplot(235)
plt.title('Fifth neuron hidden layer 1')
plt.scatter(result[:, 0], result[:, 1], c= out15.reshape(-1), cmap=cm.coolwarm)

# Sixth neuron, input layer
plt.subplot(236)
plt.title('Sixth neuron hidden layer 1')
plt.scatter(result[:, 0], result[:, 1], c= out16.reshape(-1), cmap=cm.coolwarm)
```
